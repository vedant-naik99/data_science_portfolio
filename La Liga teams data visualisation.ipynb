# La Liga Teams Points Data visualization
Data visualization plots are present in the following code to answer the following question:
* Does the form of top 3 team changes because of the competition between them.



```python
import pandas as pd
import numpy as np
import matplotlib as mlt
import matplotlib.pyplot as plt
from functools import reduce
%matplotlib notebook
```

```python
barcastats = pd.read_excel("/home/vedant/Python/laliga/barca-stats.ods", engine="odf")
barca = pd.DataFrame(barcastats, columns= ['Season','Squad','MP','W','D','L','GF','GA','GDiff','Pts'])
barca['tiewins'] = barca['D']*0.5
barca['totalwins'] = barca['W']+barca['tiewins']
barca['Barcelona Win%'] = barca['totalwins']/barca['MP']
barca['Barcelona Goal Diff'] = barca['GDiff']
barca['Barcelona Total Points'] = barca['Pts']
barcawin = pd.DataFrame(barca, columns= ['Barcelona Win%'])
barcagd = pd.DataFrame(barca, columns= ['Barcelona Goal Diff'])
barcapts = pd.DataFrame(barca, columns= ['Barcelona Total Points'])
barca
```

```python
madridstats = pd.read_excel("/home/vedant/Python/laliga/realmadrid-stats.ods", engine="odf")
madrid = pd.DataFrame(madridstats, columns= ['Season','Squad','MP','W','D','L','GF','GA','GDiff','Pts'])
madrid['tiewins'] = madrid['D']*0.5
madrid['totalwins'] = madrid['W']+madrid['tiewins']
madrid['Real Madrid Win%']= madrid['totalwins']/madrid['MP']
madrid['Real Madrid Goal Diff'] = madrid['GDiff']
madrid['Real Madrid Total Points'] = madrid['Pts']
madridwin = pd.DataFrame(madrid, columns= ['Real Madrid Win%'])
madridgd = pd.DataFrame(madrid, columns= ['Real Madrid Goal Diff'])
madridpts = pd.DataFrame(madrid, columns= ['Real Madrid Total Points'])

madrid
```

```python
atletistats = pd.read_excel("/home/vedant/Python/laliga/atleti-stats.ods", engine="odf")
atleti = pd.DataFrame(atletistats, columns= ['Season','Squad','MP','W','D','L','GF','GA','GDiff','Pts'])
atleti['tiewins'] = atleti['D']*0.5
atleti['totalwins'] = atleti['W']+atleti['tiewins']
atleti['Atletico Madrid Win%'] = atleti['totalwins']/atleti['MP']
atleti['Atletico Madrid Goal Diff'] = atleti['GDiff']
atleti['Atletico Madrid Total Points'] = atleti['Pts']
atletiwin = pd.DataFrame(atleti, columns= ['Atletico Madrid Win%'])
atletigd = pd.DataFrame(atleti, columns= ['Atletico Madrid Goal Diff'])
atletipts = pd.DataFrame(atleti, columns= ['Atletico Madrid Total Points'])
atleti
```

```python
winperba = pd.merge(barcawin, atletiwin, left_index=True, right_index=True, how='outer')
winperbar = pd.merge(winperba, madridwin, left_index=True, right_index=True, how='outer')
sea = pd.DataFrame(data={'Season': [2020, 2019, 2018,2017,2016,2015,2014,2013,2012,2011,2010,2009,2008,2007,2006,2005,2004,2003,2002,2001,2000,1999,1998,1997,1996,1995,1994,1993,1992,1991,1990,1989]})
winper = pd.merge(winperbar, sea, left_index=True, right_index=True, how='outer')
winper.set_index('Season', inplace=True)
winper.sort_index(inplace=True)
colors = ['blue','grey','red']
fig = winper.plot(color=colors)
plt.grid(True)
ax = plt.gca()
ax.set_xlabel('Season')
ax.set_ylabel('Win Percentage')
ax.set_title('Comparison of Win percentage of Atletico Madrid,Barcelona And Real Madrid', fontsize=10)
```

```python
gdba = pd.merge(barcagd, atletigd, left_index=True, right_index=True, how='outer')
gdbar = pd.merge(gdba, madridgd, left_index=True, right_index=True, how='outer')
gd = pd.merge(gdbar, sea, left_index=True, right_index=True, how='outer')
gd.set_index('Season', inplace=True)
gd.sort_index(inplace=True)
colors = ['blue','grey','red']
fig = gd.plot(color=colors)
plt.grid(True)
ax = plt.gca()
ax.set_xlabel('Season')
ax.set_ylabel('Goal Difference')
ax.set_title('Comparison of Goal Difference of Atletico Madrid,Barcelona And Real Madrid', fontsize=10)
```

```python
ptsba = pd.merge(barcapts, atletipts, left_index=True, right_index=True, how='outer')
ptsbar = pd.merge(ptsba, madridpts, left_index=True, right_index=True, how='outer')
pts = pd.merge(ptsbar, sea, left_index=True, right_index=True, how='outer')
pts.set_index('Season', inplace=True)
pts.sort_index(inplace=True)
colors = ['blue','grey','red']
fig = pts.plot(color=colors)
plt.grid(True)
ax = plt.gca()
ax.set_xlabel('Season')
ax.set_ylabel('Points')
ax.set_title('Comparison of Points of Atletico Madrid,Barcelona And Real Madrid')
```
